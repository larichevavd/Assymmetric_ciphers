```python
import socket
import datetime
import random
import pickle

sock = socket.socket()
sock.connect(('localhost', 9091))
address = sock.getsockname()[0]
sock.send(address.encode('utf-8'))
print(f'{datetime.datetime.now()} Создано соединение с сервером')
data1 = sock.recv(1024).decode('utf-8')
if data1 == 'None':
    name = input('Введите свое имя: ')
    sock.send(name.encode('utf-8'))
else:
    print(f'Добро пожаловать, {data1}')
    
g=random.randint(10, 20) # получение ключа
p=random.randint(50, 100)
a=random.randint(50, 100)
A=(g**a)%p
num = f'{A} {g} {p}'
sock.send(num.encode('utf-8'))
B = int(sock.recv(1024).decode('utf-8'))
global K
K = (B**a)%p

def to_cipher(mes): # шифрование
    mess = []
    for symbol in mes:
        mess.append(chr(ord(symbol)+K))
    return pickle.dumps(mess)

def fr_cipher(mes):
    mes_n = pickle.loads(mes)
    mess = []
    for symbol in mes_n:
        mess.append(chr(ord(symbol)-K))
    mess_n = ''.join(mess)
    return mess_n

print(f'Введите "commands", чтобы выбрать специальные возможности')
while True:
    mes = input('Enter message: ')
    sock.send(to_cipher(mes))
    if mes == 'commands':
        com = input("0: Отключение сервера\n1: Пауза\n2: Показ логов\n3: Очистка логов\n4: Очистка файла идентификации\n")
        if com == '1':
            break
        sock.send(to_cipher(com))
        if com == '0':
            break
    if mes == 'commands':
        if com == '2':
            data = sock.recv(1024).decode('utf-8')
        else:
            data = fr_cipher(sock.recv(1024))
    else:
        data = fr_cipher(sock.recv(1024))
    print(f'{datetime.datetime.now()} Recieved: ', data)
sock.close()
```
